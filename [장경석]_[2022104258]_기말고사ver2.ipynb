```python
#필요 패키지 임포트
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
```

```python
# 엑셀 파일의 경로를 지정
excel_file_path_ks2017 = "성동구립성수도서관 도서 대출목록 (2017년 05월).xlsx"
excel_file_path_ks2023 = "성동구립성수도서관 장서 대출목록 (2023년 05월).xlsx"

# 엑셀 파일을 pandas DataFrame으로 읽음
df_ks2017 = pd.read_excel(excel_file_path_ks2017)
df_ks2023 = pd.read_excel(excel_file_path_ks2023)
```

3. 2017년 5월과 2023년 5월 가장 대출이 많이 발생한 도서 상위 20권의 리스트와 그 빈도수 비교

```python
#17년, 23년 ISBN기준으로 중복된 도서 데이터 제거
duplicate_df2017 = df_2017.drop_duplicates(['ISBN'])
duplicate_df2023 = df_2023.drop_duplicates(['ISBN'])

#ISBN 그룹으로 묶어 대출건수를 합해서 책의 총 대출건수를 계산
df_2017_sum = df_2017.groupby('ISBN').sum()['대출건수'].sort_values(ascending=False)
df_2023_sum = df_2023.groupby('ISBN').sum()['대출건수'].sort_values(ascending=False)

#중복제거된 데이터프레임에 합산된 대출건수 추가
duplicate_df2017_sum = pd.merge(duplicate_df2017, df_2017_sum, on='ISBN')
duplicate_df2023_sum = pd.merge(duplicate_df2023, df_2023_sum, on='ISBN')

#대출건수로 정렬
duplicate_df2017_sum = duplicate_df2017_sum.sort_values(by='대출건수_y',ascending=False)
duplicate_df2023_sum = duplicate_df2023_sum.sort_values(by='대출건수_y',ascending=False)

#필요없는 필드 제거
duplicate_df2017_sum = duplicate_df2017_sum.drop(['대출건수_x','관리번호','저자','출판사','세트 ISBN','부가기호','권','주제분류번호','도서권수','등록일자'], axis=1)
duplicate_df2023_sum = duplicate_df2023_sum.drop(['대출건수_x','번호','저자','출판사','세트 ISBN','부가기호','권','주제분류번호','도서권수','등록일자'], axis=1)

#명확한 데이터 조회를 위해 컬럼명 변경
duplicate_df2017_sum = duplicate_df2017_sum.rename(columns={'도서명':'17년 도서명'})
duplicate_df2017_sum = duplicate_df2017_sum.rename(columns={'발행년도':'17년 발행년도'})
duplicate_df2017_sum = duplicate_df2017_sum.rename(columns={'대출건수_y':'17년 대출건수'})

duplicate_df2023_sum = duplicate_df2023_sum.rename(columns={'도서명':'23년 도서명'})
duplicate_df2023_sum = duplicate_df2023_sum.rename(columns={'발행년도':'23년 발행년도'})
duplicate_df2023_sum = duplicate_df2023_sum.rename(columns={'대출건수_y':'23년 대출건수'})

#대출건수가 많은 순서대로 상위 20개를 뽑아 데이터프레임에 저장
duplicate_df2017_sum = duplicate_df2017_sum.head(20)
duplicate_df2023_sum = duplicate_df2023_sum.head(20)

#데이터에서 대출건수 많은 순서대로 순위 지정후 데이터프레임에 추가
duplicate_df2017_sum['rank'] = duplicate_df2017_sum['17년 대출건수'].rank(ascending=False)
duplicate_df2023_sum['rank'] = duplicate_df2023_sum['23년 대출건수'].rank(ascending=False)

#순위를 인덱스로 지정
duplicate_df2017_sum.set_index(['rank'], inplace=True)
duplicate_df2023_sum.set_index(['rank'], inplace=True)

#17년, 23년 데이터프레임을 compare_df에 합쳐서 한번에 출력
#같은 행에 17년과 23년을 동시에 나타내고 싶은데 하지 못함
compare_df = pd.concat([duplicate_df2017_sum, duplicate_df2023_sum])
compare_df.reset_index(drop = True)
compare_df
```
